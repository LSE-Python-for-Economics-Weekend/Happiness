```python
Continents = ['Africa', 'Asia', ' SouthAmerica', 'Europe', 'NorthAmerica', 'Oceania']
AvgLifeladder = [4.338695, 5.278338,  6.083304,  6.115556, 6.150397, 7.310778]
plt.bar(Continents, AvgLifeladder)
plt.ylabel("Avg life Ladder")
plt.title("Avg Life Ladder in each Continent")
plt.show()
```

```python
# change the size of the figure
plt.figure(figsize=(20,15))
# plot the scatterplot
GINI_and_GDP = sns.scatterplot(x = "GINI index (World Bank estimate), average 2000-16", 
                               y = "Log GDP per capita", hue="Continent", 
                               size ="Freedom to make life choices", 
                               sizes =(-1000, 1000),
                               alpha=.8,
                               data = df, 
                               legend = 'brief')
#move the legend outside the graph
plt.legend(loc='center left', bbox_to_anchor=(1, 0.1))

#add title
GINI_and_GDP.set_title('Gini coefficient and GDP',fontsize = 35)
```

```python
# change the size of the figure
plt.figure(figsize=(20,15))
#plot the scatter plot
GINI_and_Life = sns.scatterplot(x = "GINI index (World Bank estimate), average 2000-16",
                                y="Life Ladder",
                                size ="Freedom to make life choices", 
                                sizes =(-2000, 2000),
                                hue="Continent", 
                                data=df,
                                legend='brief')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.1))
#add title
GINI_and_Life.set_title('Gini coefficient and Life Ladder', fontsize = 35)
```
