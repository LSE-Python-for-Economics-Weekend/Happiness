```python
import seaborn as sns
%matplotlib inline
df = pd.read_csv('happiness_with_continent(2019).csv.txt')
df = df[["Log GDP per capita","Life Ladder",'Continent']].dropna(how='any',axis=0)
fig = plt.figure(figsize=(10, 5))

x = df['Log GDP per capita']
y = np.log(df["Life Ladder"]) 

color_dict = {'Africa':'green',
                  'Asia':'blue',
                  'Europe': 'orange',
                  'Oceania': 'purple',
                  'South America': 'red',
                  'North America' : 'brown'}


# dictionary = {'':'', '':''}
# dictioanry = dict([('Africa', 'brown'), ('Asia', 'Green')])
for continent in df['Continent'].unique():
    tempdf = df[df['Continent'] == continent]
    tempdf['Life Ladder'] = np.log(tempdf['Life Ladder'])
    tempdf['Life Ladder'].legend_ = None
    plt.scatter('Log GDP per capita', 'Life Ladder', data = tempdf,  c= color_dict[continent], label= None)
    
    m, b = np.polyfit(tempdf['Log GDP per capita'], tempdf['Life Ladder'], 1)
    plt.plot(tempdf['Log GDP per capita'], m*tempdf['Log GDP per capita'] + b, color = color_dict[continent])
    

m, b = np.polyfit(x, y ,1)
plt.plot(x, m*x + b,color='black')

g =sns.scatterplot(x=x, y=y, hue="Continent", data=df, legend='brief')
plt.xlabel("Log GDP per capita")  
plt.ylabel("Log Life Ladder")
plt.title('Correlation between GDP per capita and Life Ladder')
plt.legend(loc='lower right',fontsize='9.2')
```
